```python
# data cleaning, html parsing, and date/time management libraries
import numpy as np
import pandas as pd
import requests
import html5lib
import time
import datetime as dt
import pytz

# Selenium & BeautifulSoup scraping/web browser automation libraries
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.keys import Keys
from bs4 import BeautifulSoup


# plotly plotting libraries
from plotly.offline import init_notebook_mode, plot, iplot
from plotly.graph_objs import *
init_notebook_mode(connected=True)

# virtual display libraries for headless scraping
from pyvirtualdisplay import Display
display = Display(visible=0,size=(1280,1024))
display.start()
```

```python
url = 'https://www.twitch.tv/directory'
driver = webdriver.Chrome()
driver.get(url)
```

```python
def load_more_results(elem):
    elem.send_keys(Keys.PAGE_DOWN)
    time.sleep(3)
    elem.send_keys(Keys.PAGE_DOWN)
    time.sleep(3)
    soup = BeautifulSoup(driver.page_source,'html5lib')
    return soup

def list_views(soup):
    return [entry for entry in soup.findAll('div',{'data-target':'directory-container'})[0].text.split('viewers')[:-1]]
```

```python
soup = BeautifulSoup(driver.page_source,'html5lib')
elem = driver.find_element_by_partial_link_text("Fortnite")
time.sleep(1)
results = list_views(soup)
last_game = results[~0]
```

```python
#scrape away!
count = 1
R = 0
while R < len(list_views(load_more_results(elem))):
    print 'Grabbing batch {}: > {} Games'.format(count,R)
    count += 1
    soup = load_more_results(elem)
    results = list_views(soup)
    print results[~0]
    R = len(results)
```

```python
results_dic={}
for i in range((len(soup.findAll('div',{'class':'tw-mg-t-05'})))):
    results_dic[soup.findAll('div',{'class':'tw-mg-t-05'})[i].h3['title']] = \
        soup.findAll('div',{'class':'tw-mg-t-05'})[i].p['title']
    
```

```python
results_dic
```

```python
results_df = pd.Series(results_dic).reset_index().rename(columns={'index':'Game',0:'Views'})
results_df['Views'] = results_df['Views'].map(lambda x: x.replace(',','').split('viewer')[0].strip())
results_df['Views'] = results_df['Views'].astype(int)
results_df = results_df.sort_values(by=['Views'],ascending=[0])
results_df.head()
```

```python
import datetime as dt
import pytz

scrape_time =  dt.datetime.now(pytz.timezone('US/Eastern')).replace(second=0, microsecond=0)
```

```python
iplot(Figure(data=[Bar(x=results_df[results_df['Game'].isin(results_df['Game'][:25])]['Game'].map(lambda x: "<i><b>{}</i></b>".format(x)),
                   y=results_df[results_df['Game'].isin(results_df['Game'][:25])]['Views'],
                   opacity=.8,
                   marker={'color':'#C1F8BE'})],
                   layout=Layout(title='<b>Twitch TV</b> Viewers By Game<br>Top 25 Most Popular Games<br><b>{D} EST</b>'.format(D=str(scrape_time)[:~5]),
                                 width=1000, height=800,
                                 margin={'b':160},
                             xaxis={'autotick':True,'tickangle':30})
            )
     )
```

```python
total_current_views = results_df['Views'].cumsum().max()
results_df['Views Pct.'] = results_df['Views'].map(lambda x: round(float(x)/float(total_current_views),3)*100)
results_df['Views Cumulative Pct.'] = results_df['Views'].cumsum().map(lambda x: round(float(x)/float(total_current_views),3)*100)
```
